```python
import requests
response =requests.get("https://raw.githubusercontent.com/mattiabusi/DATASET-OSPEDALI/refs/heads/main/csvjson%20(1).json")
data=response.json()
```

ELENCO DEGLI OSPEDALI RAGGRUPPATI PER CITTÁ

```python
# Filtra gli ospedali con un rating valido
valutazioni_valide = [ospedale for ospedale in data if 'Hospital overall rating' in ospedale and ospedale['Hospital overall rating'] not in [None, "Not Available"]]

# Converte la valutazione in un numero
for ospedale in valutazioni_valide:
    ospedale['Hospital overall rating'] = float(ospedale['Hospital overall rating'])

# Ordina per valutazione in ordine decrescente
valutazioni_valide.sort(key=lambda x: x['Hospital overall rating'], reverse=True)

# Seleziona i primi 10
top_10_ospedali = valutazioni_valide[:10]

# Stampa i risultati
print("\nTop 10 ospedali con valutazione maggiore:")
for indice, ospedale in enumerate(top_10_ospedali, start=1):
    print(f"{indice}. {ospedale['Hospital Name']} ({ospedale['City']}, {ospedale['State']}): {ospedale['Hospital overall rating']}")
```

ELENCO OSPEDALI CON CITTÁ E STATO

```python
# Stampa dell'elenco degli ospedali con città e stato
print("\nElenco degli ospedali con città e stato:")
for ospedale in data:
    if 'Hospital Name' in ospedale and 'City' in ospedale and 'County Name' in ospedale:
        print(f"- {ospedale['Hospital Name']} ({ospedale['City']}, {ospedale['County Name']})")
```

CONTA GLI OSPEDALI PER OGNI STATO

```python
# Conta il numero di ospedali per stato
ospedali_per_stato = {}
for ospedale in data:
    if 'State' in ospedale:
        stato = ospedale['State']
        if stato not in ospedali_per_stato:
            ospedali_per_stato[stato] = 0
        ospedali_per_stato[stato] += 1

# Stampa il risultato
print("\nNumero di ospedali per stato:")
for stato, conteggio in ospedali_per_stato.items():
    print(f"{stato}: {conteggio}")
```

OSPEDALI CHE OFFRONO SERVIZI DI EMERGENZA

```python
# Filtra gli ospedali che offrono servizi di emergenza
ospedali_emergenza = [ospedale for ospedale in data if ospedale.get("Emergency Services") == "TRUE"]

# Stampa gli ospedali con servizi di emergenza
print("\nOspedali che offrono servizi di emergenza:")
for ospedale in ospedali_emergenza:
    if 'Hospital Name' in ospedale and 'City' in ospedale and 'State' in ospedale:
        print(f"- {ospedale['Hospital Name']} ({ospedale['City']}, {ospedale['State']})")
```

VALUTAZIONE MEDIA PER OGNI TIPO DI OSPEDALE

```python
# Calcola la valutazione media per tipo di ospedale
tipi_ospedali = {}
for ospedale in data:
    if 'Hospital Type' in ospedale and 'Hospital overall rating' in ospedale:
        tipo_ospedale = ospedale['Hospital Type']
        valutazione = ospedale['Hospital overall rating']

        # Verifica se la valutazione è valida prima di procedere
        if valutazione not in [None, "Not Available"]:
            try:
                # Converte la valutazione in numero
                valutazione = float(valutazione)

                # Aggiungi la valutazione alla somma totale per tipo di ospedale
                if tipo_ospedale not in tipi_ospedali:
                    tipi_ospedali[tipo_ospedale] = {'somma_valutazioni': 0, 'conteggio': 0}
                tipi_ospedali[tipo_ospedale]['somma_valutazioni'] += valutazione
                tipi_ospedali[tipo_ospedale]['conteggio'] += 1
            except ValueError:
                # Ignora i valori che non possono essere convertiti
                print(f"Valore non valido per la valutazione: {valutazione} (per {ospedale['Hospital Name']})")

# Stampa
print("\nValutazione media per tipo di ospedale:")
for tipo_ospedale, valori in tipi_ospedali.items():
    valutazione_media = valori['somma_valutazioni'] / valori['conteggio'] if valori['conteggio'] > 0 else 0
    print(f"{tipo_ospedale}: {valutazione_media:.2f}")
```

FILTRAGGIO OSPEDALI PER CALIFORNIA

```python
# Stato California da filtrare
stato_da_filtrare = "CA"

# Filtra gli ospedali per stato
ospedali_filtrati = [ospedale for ospedale in data if ospedale.get("State") == stato_da_filtrare]

# Stampa gli ospedali
print(f"\nOspedali nello stato {stato_da_filtrare}:")
for ospedale in ospedali_filtrati:
    if 'Hospital Name' in ospedale and 'City' in ospedale:
        print(f"- {ospedale['Hospital Name']} ({ospedale['City']})")
```

GRAFICO OSPEDALI PER TIPO

```python
import matplotlib.pyplot as plt

# Conta gli ospedali per tipo
tipi_ospedali = {}
for ospedale in data:
    if 'Hospital Type' in ospedale:
        tipo_ospedale = ospedale['Hospital Type']
        tipi_ospedali[tipo_ospedale] = tipi_ospedali.get(tipo_ospedale, 0) + 1

# Crea il grafico a barre
plt.figure(figsize=(12, 6))
plt.bar(tipi_ospedali.keys(), tipi_ospedali.values(), color='lightcoral', edgecolor='black')
plt.title('Distribuzione degli Ospedali per Tipo')
plt.xlabel('Tipo di Ospedale')
plt.ylabel('Numero di Ospedali')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

GRAFICO OSPEDALI PER STATO

```python
import matplotlib.pyplot as plt

# Conta il numero di ospedali per stato
conteggio_per_stato = {}
for ospedale in data:
    if 'State' in ospedale:
        stato = ospedale['State']
        conteggio_per_stato[stato] = conteggio_per_stato.get(stato, 0) + 1

# Crea il grafico a barre orizzontale
plt.figure(figsize=(12, 8))
plt.barh(list(conteggio_per_stato.keys()), list(conteggio_per_stato.values()), color='lightseagreen')
plt.title('Numero di Ospedali per Stato')
plt.xlabel('Numero di Ospedali')
plt.ylabel('Stato')
plt.show()
```

GRAFICO OSPEDALI CON E SENZA SERVIZI DI EMERGENZA

```python
import matplotlib.pyplot as plt

# Conta ospedali con e senza servizi di emergenza
conteggio_emergenza = {'TRUE': 0, 'FALSE': 0}
for ospedale in data:
    if 'Emergency Services' in ospedale:
        conteggio_emergenza[ospedale['Emergency Services']] += 1

# Crea un grafico a torta
plt.figure(figsize=(7, 7))
plt.pie(conteggio_emergenza.values(), labels=conteggio_emergenza.keys(), autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Percentuale di Ospedali con e senza Servizi di Emergenza')
plt.show()
```
